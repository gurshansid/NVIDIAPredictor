```python
nvidia = nvidia.history(period="max")
```

```python
del nvidia ["Dividends"]
del nvidia ["Stock Splits"]
```

```python
nvidia["Tomorrow"] = nvidia["Close"].shift(-1)
```

```python
nvidia
```

```python
nvidia["Target"] = (nvidia["Tomorrow"] > nvidia["Close"]).astype(int)
```

```python
nvidia
```

```python
nvidia = nvidia.loc["2000-01-01":].copy()
```

```python
nvidia
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, min_samples_split=300, random_state=1)
train = nvidia.iloc[:-100]
test = nvidia.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

predictions = model.predict(test[predictors])
```

```python
import pandas as pd

predictions = pd.Series(predictions, index=test.index)
```

```python
predictions
```

```python
precision_score(test["Target"], predictions)
```

```python
all_values = pd.concat([test["Target"], predictions], axis=1)
```

```python
all_values.plot()
```

```python
def predictor(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    predictions = model.predict(test[predictors])
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    all_values = pd.concat([test["Target"], predictions], axis=1)
    return all_values
```

```python
def backtest(data, model, predictors, start=2500, step=250):
  results = []

  for i in range(start, data.shape[0], step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predictor(train, test, predictors, model)
    results.append(predictions)
  return pd.concat(results)
```

```python
predictions = backtest(nvidia, model, predictors)
```

```python
rolling_average_horizons = [3, 5, 62, 250, 1000]
new_predictors = []

for horizon in rolling_average_horizons:
  rolling_averages = nvidia.rolling(horizon).mean()

  ratio_column = f"Close_Ratio_{horizon}"
  nvidia[ratio_column] = nvidia["Close"] / rolling_averages["Close"]

  trend_column = f"Trend_{horizon}"
  nvidia[trend_column] = nvidia.shift(1).rolling(horizon).sum()["Target"]

  new_predictors += [ratio_column, trend_column]
```

```python
nvidia = nvidia.dropna()
```

```python
nvidia
```

```python
model = RandomForestClassifier(n_estimators=3000, min_samples_split=200, random_state=1)
```

```python
def predictor(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    predictions = model.predict_proba(test[predictors]) [:,1]
    predictions[predictions >= .6] = 1
    predictions[predictions < .6] = 0
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    all_values = pd.concat([test["Target"], predictions], axis=1)
    return all_values
```

```python
predictions = backtest(nvidia, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
